```python
import statsmodels
import scipy as sc
import numpy as np
import pandas as pd
import seaborn as sns 

%matplotlib inline
```

```python
def write_answer(task,ans):
    with open("bio_" + str(task)+".txt", "w") as fout:
        fout.write(str(ans))
```

```python
gene_data = pd.read_csv('gene_high_throughput_sequencing.csv')
```

```python
gene_data.drop('Patient_id', axis = 1, inplace = True)
```

```python
gene_data.head()
```

```python
gene_data.Diagnosis.value_counts()
```

```python
normal = gene_data.loc[gene_data.Diagnosis == 'normal'].drop('Diagnosis', axis = 1)
early_np = gene_data.loc[gene_data.Diagnosis == 'early neoplasia'].drop('Diagnosis', axis = 1)
cancer = gene_data.loc[gene_data.Diagnosis == 'cancer'].drop('Diagnosis', axis = 1)
```

```python
normal.LOC643837.mean()
```

```python
p_values = pd.DataFrame(index = gene_data.columns[1:], columns = ['NORM_EARLYnp', 'EARLYnp_CANCER'])

for gene in p_values.index:
    p_values['NORM_EARLYnp'][gene] = sc.stats.ttest_ind(normal[gene], early_np[gene], equal_var = False).pvalue
    p_values['EARLYnp_CANCER'][gene] = sc.stats.ttest_ind(early_np[gene], cancer[gene], equal_var = False).pvalue
    
p_values
```

```python
ans1_1 = (p_values.NORM_EARLYnp < 0.05).sum()
write_answer(1.1, ans1_1)
```

```python
ans1_2 =  (p_values.EARLYnp_CANCER < 0.05).sum()
write_answer(1.2, ans1_2)
```

## Часть 2

```python
import statsmodels.stats.multitest as smm
```

```python
reject, p_corrected_holm,  a1, a2 = smm.multipletests(p_values.NORM_EARLYnp, alpha = 0.025, method='holm')

p_values['NORM_EARLYnp_holm'] = p_corrected_holm
p_values['NORM_EARLYnp_reject_holm'] = reject
```

```python
reject, p_corrected_holm,  a1, a2 = smm.multipletests(p_values.EARLYnp_CANCER, alpha = 0.025, method='holm')

p_values['EARLYnp_CANCER_holm'] = p_corrected_holm
p_values['EARLYnp_CANCER_reject'] = reject
```

```python
p_values.head()
```

```python
def fold_change (control, treatment):
    if treatment > control:
        return abs(treatment/control)
    return abs(control/treatment)
```

```python
reject_true1 = p_values[p_values.NORM_EARLYnp_reject_holm == True].index

reject_true2 = p_values[p_values.EARLYnp_CANCER_reject == True].index
```

```python
ans2_1 = 0
for gene in reject_true1:
    fc = fold_change(normal[gene].mean(), early_np[gene].mean())
    print (fc)
    if fc > 1.5:
        ans2_1 +=1
        print (gene)
```

```python
write_answer(2.1, ans2_1)
```

```python
ans2_2 = 0
for gene in reject_true2:
    fc = fold_change(early_np[gene].mean(), cancer[gene].mean())
    if fc > 1.5:
        ans2_2 +=1
```

```python
write_answer(2.2, ans2_2)
```
